# Loading and cleaning the data

```python
# importing the necessary library for data analysis and cleaning
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Loading the csv data and assigning it to df 
df=pd.read_csv("abcsales2.csv")
```

```python
# Displaying first 5 rows of the dataset
df.head()
```

```python
# Defining the shape and size of the data
df.shape
```

```python
# Defining the type of data
df.info()
```

```python
# changing datatype of Date to date type from object
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
# Checking for null value
df.isnull().sum()
```

```python
# Dropping the null values
df.dropna(inplace=True)
```

```python
# Rechecking is null value are dropped successfully
df.isnull().sum()
```

```python
#Displaying 10 dataset from top for better understanding of data
df.head(10)
```

```python
#Displaying the data of last 10 columns
df.tail(10)
```

# The data is cleaned succesfully 

# Exploratory data analysis

```python
gender=df['Gender'].value_counts()
plt.pie(gender,labels=gender.index,explode=[0,0.05],autopct='%1.1f%%')
plt.show()
```

```python
# From given figure we can conclude that women are slightly more active in shopping
```

```python
customer=df['Customer type'].value_counts()
plt.pie(customer,explode=[0,0.05],labels=customer.index,autopct='%1.1f%%')
plt.show()
```

```python
 # Within the shop,Often normal buyer who are not member visits more frequently than members. 
```

```python
#Plotting annual sales of products within the organization
product = sns.countplot(x="Product line",data=df,width=0.6)
plt.xticks(rotation=40, ha="right")
product.bar_label(product.containers[0])
plt.title("Product line")
```

```python
# Although products consumption rate is similar.oil,masalas and food beverages are in high demand as compared to other
```

```python
#Payment Gateaway analysis
payment = sns.countplot(x="Payment",data=df,width=0.6)
plt.xticks(rotation=40, ha="right")
payment.bar_label(payment.containers[0])
plt.title("Payment Gateaway")
```

```python
# Although digital payment method is available People prefer cash more than other payment.
```

```python
#Plotting shopping count accroding to gender
product = sns.countplot(x="Gender",data=df,width=0.6)
plt.xticks(rotation=40, ha="right")
product.bar_label(product.containers[0])
plt.title("Product line")
plt.show()
```

```python
# Calculating total spend on shopping according to gender
# Group by 'Gender' and sum the 'Total' column
gender_totals = df.groupby('Gender')['Total'].sum().reset_index()

# Plot the bar chart
plt.figure(figsize=(8, 5))
plt.bar(gender_totals['Gender'], gender_totals['Total'])
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.title('Sum of Total Amount by Gender')
plt.xticks(rotation=40, ha="right")

# Annotate the bars with the total amount
for index, value in enumerate(gender_totals['Total']):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
# As women are heigher in shopping more money is spent by women rather than men.
```

```python
# Group by 'Gender' and sum the 'Total' column
gender_totals = df.groupby('Customer type')['Total'].sum().reset_index()

# Plot the bar chart
plt.figure(figsize=(8, 5))
plt.bar(gender_totals['Customer type'], gender_totals['Total'])
plt.xlabel('Customer type')
plt.ylabel('Total Amount')
plt.xticks(rotation=40, ha="right")
plt.title('Sum of Total Amount by Member Type')

# Annotate the bars with the total amount
for index, value in enumerate(gender_totals['Total']):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
# As normal nom-member are active in shopping, The transaction amount is more in case of Non-Members than that of members.
```

```python
# Plotting the sales quaterly
# Extract the quarter and year
df['Quarter'] = df['Date'].dt.to_period('Q')

# Group by quarter and sum sales
quarterly_sales = df.groupby('Quarter')['Total'].sum().reset_index()

# Plot the results
plt.figure(figsize=(5, 3))
plt.plot(quarterly_sales['Quarter'].astype(str), quarterly_sales['Total'], marker='o')
plt.title('Sales by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# The sales is High in end of Quater i.e Quater 4.The sales comparatively decreases and comeback to normal form
```

```python
# Extract the quarter and year
df['Quarter'] = df['Date'].dt.to_period('Q')

# Group by quarter and sum sales
quarterly_sales = df.groupby('Quarter')['Total'].sum().reset_index()

# Convert 'Quarter' to string for better visualization
quarterly_sales['Quarter'] = quarterly_sales['Quarter'].astype(str)

# Plot the results
plt.figure(figsize=(6, 4))
plt.bar(quarterly_sales['Quarter'], quarterly_sales['Total'], color='skyblue')
plt.title('Total Sales by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the total sales amount on top of each bar
for index, value in enumerate(quarterly_sales['Total']):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom')

plt.show()
```

```python
# The sales amount in higher in Quater end of 2023.
```

```python
# Extract the month and year
df['Month'] = df['Date'].dt.to_period('M')

# Group by month and sum sales
monthly_sales = df.groupby('Month')['Total'].sum().reset_index()

# Convert 'Month' to string for better visualization
monthly_sales['Month'] = monthly_sales['Month'].astype(str)

# Plot the results
plt.figure(figsize=(12, 8))
plt.bar(monthly_sales['Month'], monthly_sales['Total'], color='lightgreen')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the total sales amount on top of each bar
for index, value in enumerate(monthly_sales['Total']):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom')

plt.show()
```

```python
# The sales reaches maximum in August .
```

```python
# Extract the quarter and year
df['Quarter'] = df['Date'].dt.to_period('M')

# Group by quarter and sum sales
quarterly_sales = df.groupby('Quarter')['Total'].sum().reset_index()

# Plot the results
plt.figure(figsize=(5, 3))
plt.plot(quarterly_sales['Quarter'].astype(str), quarterly_sales['Total'], marker='o')
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# The sales gradually increases in August and gradually comeback to its mean position.The sales reaches heigher point in September,October and November 
```

```python
#Generating total sales amount according to mode of payment
# Group by mode of payment and sum sales
payment_mode_sales = df.groupby('Payment')['Total'].sum().reset_index()

# Plot the results
plt.figure(figsize=(10, 6))
plt.bar(payment_mode_sales['Payment'], payment_mode_sales['Total'], color='coral')
plt.title('Total Sales by Payment Mode')
plt.xlabel('Payment Mode')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the total sales amount on top of each bar
for index, value in enumerate(payment_mode_sales['Total']):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom')

plt.show()
```

```python
# High revenue is generated from cash
```

```python
# Group by product line and sum tax
product_line_tax = df.groupby('Product line')['Tax 5%'].sum().reset_index()

# Plot the results
plt.figure(figsize=(8, 4),dpi=100)
plt.bar(product_line_tax['Product line'], product_line_tax['Tax 5%'], color='lightblue')
plt.title('Total Tax Amount by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Tax Amount')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# More Tax is generated from Food grains
```

```python
# Group by product line and gender, and sum sales amount
product_line_gender_sales = df.groupby(['Product line', 'Gender'])['Total'].sum().unstack().fillna(0)

# Plot the results
product_line_gender_sales.plot(kind='bar', figsize=(12, 6), color=['skyblue', 'lightcoral'])
plt.title('Total Sales Amount by Product Line and Gender')
plt.xlabel('Product Line')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Gender')
plt.show()
```

```python
# Males buy Food grains more than other goods while women buys frozen and poultry foods compratively more than man.
```
